```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
import json
from tqdm.auto import tqdm, trange
```

```python
template = 'https://fennougrica.kansalliskirjasto.fi/discover?rpp=10&etal=0&group_by=none&page={}&filtertype_0=language&filter_relational_operator_0=equals&filter_0=fi%3DErsä%7Cen%3DErzyan%7Cru%3DЭрзянский+язык%7C'
```

```python
base_url = 'https://fennougrica.kansalliskirjasto.fi'
```

```python
book_urls = []

for page in trange(1, 15+1):
    soup = BeautifulSoup(requests.get(template.format(page)).text)
    for ad in soup.findAll('div', {'class': "artifact-description"}):
        book_urls.append(base_url + ad.find('a')['href'])
print(len(book_urls))
```

```python
print(book_urls[15])
```

```python
from urllib.request import urlretrieve
import os
```

```python
BOOK_ROOT = 'fennougrica/'
```

```python
all_found = []
```

```python
for i, book_url in enumerate(tqdm(book_urls)):
    soup = BeautifulSoup(requests.get(book_url).text)
    file_url = base_url + soup.find('div', {'class': 'file-metadata'}).find('a')['href']
    title = soup.find('div', {'id': 'ds-content'}).find('h1').text
    
    file_name = '{}_{}'.format(i, file_url.split('?')[0].split('/')[-1])
    all_found.append([file_url, file_name, title])
    print(file_url)
    print(file_name)
    print(title)
    if os.path.exists(BOOK_ROOT + file_name):
        continue
    urlretrieve(file_url, filename=BOOK_ROOT + file_name)
```
