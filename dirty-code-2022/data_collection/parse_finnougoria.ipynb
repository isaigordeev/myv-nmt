```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'https://www.finnougoria.ru'
```

```python
list_pages = [
    'https://www.finnougoria.ru/logos/proza/2092/',
    'https://www.finnougoria.ru/logos/proza/2051/',
    'https://www.finnougoria.ru/logos/proza/2499/',
    'https://www.finnougoria.ru/logos/proza/1715/',
    'https://www.finnougoria.ru/logos/proza/2484/',
    'https://www.finnougoria.ru/logos/proza/1839/'
]
```

```python
results = []

for list_url in list_pages:
    soup = BeautifulSoup(requests.get(list_url).text)
    table = soup.find('div', {'class': 'catalog-section-list'})
    while table.name != 'table':
        table = table.previous_sibling
    for a in table.findAll('a'):
        url = base_url + a['href']
        print(url)
        soup = BeautifulSoup(requests.get(url).text)
        cat = soup.find('div', {'class': 'catalog-element'})
        sections = []
        lines = []
        for child in cat.children:
            if child.name == 'hr':
                sections.append(lines)
                lines = []
                continue
            if child.name == 'p' and '<em>' not in str(child):
                lines.append(child.text)
            elif child.name is None:
                lines.append(str(child))
        sections.append(lines)
        assert len(sections) <= 2
        print([len(s) for s in sections])
        results.append({
            'url': url,
            'sections': sections
        })
```

```python
import razdel
```

```python
parallel_sections = []
for res in results:
    if len(res['sections']) == 1:
        continue
    sent_lists = [
        [sent.text.strip().lstrip('***').strip() for sent in razdel.sentenize(' '.join(sec))]
        for sec in res['sections']
    ]
    parallel_sections.append({
        'url': res['url'],
        'rus': [s for s in sent_lists[1] if s],
        'myv': [s for s in sent_lists[0] if s],
    })
```

```python
import pandas as pd
```

```python
d = pd.DataFrame(parallel_sections)
d['rus_len'] = d.rus.apply(len)
d['myv_len'] = d.myv.apply(len)
d
```
