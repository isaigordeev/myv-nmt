````python
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer, LTChar

page_id = 0

items = []

for page_layout in extract_pages("russian-moksha-erzya-dictionary.pdf"):
    page_id += 1
    if page_id < 11:
        continue
    if page_id  > 11:
        break
    for element in page_layout:
        if isinstance(element, LTTextContainer):
            for text_line in element:
                #print(text_line.bbox)
                #print(text_line.get_text())
                text = []
                for c in text_line:
                    t = c.get_text()
                    if isinstance(c, LTChar) and c.fontname == 'LPBCNO+TimesNewRomanPSMT':  
                        # upper index or something
                        if t in '123456789':
                            continue
                        text.append(t)
                    elif isinstance(c, LTChar) and 'LPBDBD+TT5A57o00' in c.fontname:
                        # italics
                        text.append(f'<i>{t}</i>')
                    elif isinstance(c, LTChar) and 'LPBCPB+TT5A55o00' in c.fontname:
                        # italics
                        text.append(f'<b>{t}</b>')
                    else:
                        text.append(t)
                text = ''.join(text).strip().replace('</i><i>', '')
                text = ''.join(text).strip().replace('</b><b>', '')
                # text = re.sub('<i>([\d.,-;()\s\/]+)</i>', '\g<1>', text)
                
                items.append([text, text_line.bbox])
````

````python
import re

def get_prhases(raw):
    result = []
    # join word breaks:
    for item in raw:
        if result and result[-1].endswith('-'):
            result[-1] = result[-1][:-1] + item
        else:
            result.append(item)
    # join other line breaks
    result_text = ' '.join(result)
    # remove text in brackets
    result_text = re.sub('\([^()]*\)', ' ', result_text)
    # remove text in italics
    result_text = re.sub('<i>[^<>]*</i>', ' ', result_text)
    # remove text in bold
    result_text = re.sub('<b>[^<>]*</b>', ' ', result_text)
    
    result_text = re.sub('\s+', ' ', result_text)
    
    # skip the text if it refers to another entry
    if result_text.startswith('см.'):
        return []
    # skip the text if it has unbalanced parentheses (error)
    if result_text.count('(') != result_text.count(')'):
        return []
    
    # split into variations
    translations = []
    for section in result_text.split(';'):
        section = section.strip(' ').strip('.').strip(' ').strip('.')
        section = re.sub('^\d+\.?\s+', ' ', section)
        for phrase in section.split(','):
            phrase = phrase.strip(' ').strip('.').strip(' ').strip('.')
            if phrase:
                translations.append(phrase)
    return translations
````

# Multi-page extraction

````python
epsilon = 2
````

````python
from tqdm.auto import tqdm, trange
````

````python
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer, LTChar, LTTextLineHorizontal

page_id = 0

rus_items = []
moksha = []
erzya = []

for page_layout in tqdm(extract_pages("russian-moksha-erzya-dictionary.pdf")):
    page_id += 1
    if page_id < 8:
        continue
    if page_id > 531:
        break
    items = []
    for element in page_layout:
        if isinstance(element, LTTextContainer):
            for text_line in element:
                if not isinstance(text_line, LTTextLineHorizontal):
                    # some lines are instead empty characters
                    continue
                #print(text_line.bbox)
                #print(text_line.get_text())
                text = []
                for c in text_line:
                    t = c.get_text()
                    if isinstance(c, LTChar) and c.fontname == 'LPBCNO+TimesNewRomanPSMT':  
                        # upper index or something
                        if t in '123456789':
                            continue
                        text.append(t)
                    elif isinstance(c, LTChar) and 'LPBDBD+TT5A57o00' in c.fontname:
                        # italics
                        text.append(f'<i>{t}</i>')
                    elif isinstance(c, LTChar) and 'LPBCPB+TT5A55o00' in c.fontname:
                        # italics
                        text.append(f'<b>{t}</b>')
                    else:
                        text.append(t)
                text = ''.join(text).strip().replace('</i><i>', '')
                text = ''.join(text).strip().replace('</b><b>', '')
                # text = re.sub('<i>([\d.,-;()\s\/]+)</i>', '\g<1>', text)
                
                items.append([text, text_line.bbox])

    # extend russian items
    offset = len(rus_items)
    new_items = 0
    for word, rect in sorted(items, key=lambda x: (-x[1][1], x[1][0])):
        if rect[0] < 40:
            if word == '(cid:2)':
                continue
            if not word.strip().startswith('<b>'):  # probably a comment
                continue
            new_items += 1
            rus_items.append([word, rect])
            moksha.append([])
            erzya.append([])

    # extend moksha_items       
    for word, rect in sorted(items, key=lambda x: (-x[1][1], x[1][0])):
        if 100 < rect[0] < 150:
            prev_idx = None
            for idx, rus_item in enumerate(rus_items[offset:]):
                if rus_item[1][1] + epsilon < rect[1]:
                    break
                prev_idx = idx
            if prev_idx is not None:
                moksha[offset + prev_idx].append(word)
            elif offset:
                moksha[offset - 1].append(word)
                
    # extend erzya items
    for word, rect in sorted(items, key=lambda x: (-x[1][1], x[1][0])):
        if 220 < rect[0] < 300:
            prev_idx = None
            for idx, rus_item in enumerate(rus_items[offset:]):
                if rus_item[1][1] + epsilon < rect[1]:
                    break
                prev_idx = idx
            if prev_idx is not None:
                erzya[offset + prev_idx].append(word)
            elif offset:
                erzya[offset - 1].append(word)
````

````python
def fix_ru(text):
    for sep in ['<', ':', '(', '1', '2', '3', '4', '5', '.']:
        text = text.split(sep)[0].strip()
    
    # on probable errors, return nothing
    if '.' in text:
        return ''
    # probably, moksha words glued together to the Russian one
    if len(text.split()) > 2: 
        return text.split()[0]
    return text
````
